## Matplotlib
> Matplotlib is a Python 2D plotting library which produces publication quality figures in a variety of hardcopy formats and interactive environments across platforms.  
Matplotlib tries to make easy things easy and hard things possible.  
For simple plotting the pyplot module provides a MATLAB-like interface

https://matplotlib.org

Importing matplotlib looks like this

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### Two interfaces
There are two ways to interact with Matplot lib: a Matlab style and an object oriented style interface.

See Ch 4 in Python Data Science Handbook, Jake VanderPlas

- Two Interfaces for the Price of One, pp. 222
- Matplotlib Gotchas, pp. 232

### Matlab style interface

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

Adding decorations to the plot is done by repeatatly calling functions on the imported `plt` module. All calls within the cell will be applied to the current figure and axes.

```python
plt.plot(x, np.sin(x), '-', label='sine')
plt.plot(x, np.cos(x), '--', label ='cosine')

plt.xlim([0, 10])
plt.ylim([-1, 1])
plt.xlabel('x')
plt.ylabel('amplitude')
plt.title('My sines')
plt.grid()
plt.legend();
```

### Object oriented interface 

With this interface, you first create a figure and an axes object, then call their methods to change the plot.

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.sin(x), '-')
ax.plot(x, np.cos(x), '--');
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.sin(x), '-', label='sine')
ax.plot(x, np.cos(x), '--', label ='cosine')

ax.set(xlim=[0, 10], ylim=[-1, 1],
       xlabel='x', ylabel='amplitude',
       title='My sines');
ax.grid()
ax.legend();
```

### Save to file 
With the figure object at hand, we can save to file

```python
fig.savefig('sines.pdf')
```

## Plotting with pandas
> We use the standard convention for referencing the matplotlib API ...
> We provide the basics in pandas to easily create decent looking plots. 

https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html

Let's load the auto_mpg dataset

```python
data = pd.read_csv('auto_mpg_df.csv')
```

```python
# check of csv file created in 1st notebook...
print(data.isnull().values.any())
data.head()
```

Plotting all columns, works, but does not provide a lot of insight.

```python
data.plot();
```

Let's look at the mpg distribution (a histogram)

```python
data['mpg'].plot.hist();
```

How many samples do we have from each origin?

```python
data.origin.value_counts()
```

Notice that we accessed the origin column with dot notation. This can be done whenever the column name is 'nice' enough to be a python variable name.

Do we have similar mpg in each origin?  
Plotting two histograms for each origin side by side directly form the dataframe:

```python
data.hist(column='mpg', by='origin')
```

From the above plots it appears that each origin location is somewhat different...e.g. the modes are different, Origin 1 & 2 are right skewed, etc.

To format this plot, we can work on the axes (array) that is returned by the plot call. We use Matplotlib object oriented interface methods to do this

```python
axs = data.hist(column='mpg', by='origin')
axs[0,0].set(title='origin-1', ylim=[0, 45])
axs[0,1].set(title='origin-2', ylim=[0, 45])
axs[1,0].set(title='origin-3', ylim=[0, 45]);
```

Is mpg and displacement (disp) correlated? Maybe it is different with respect to origin?  
Let's have a look with a scatter plot.

```python
data.plot.scatter('mpg', 'disp', c='origin', colormap='winter');
```

According to:  
https://stackoverflow.com/questions/43578976/pandas-missing-x-tick-labels

the missing x-labels are a pandas bug.

Workaraound is to create axes prior to calling plot

```python
fig, ax = plt.subplots()
data.plot.scatter('mpg', 'disp', c='origin', colormap='viridis', ax=ax);
```

It is a bit annoying that there is a colorbar, we know origin is categorical.

One way to avoid the colorbar is to loop over the categories and assign colors based on the category.

See:
https://stackoverflow.com/questions/26139423/plot-different-color-for-different-categorical-levels-using-matplotlib

```python
colors = {1: 'tab:blue', 2: 'tab:orange', 3: 'tab:red'}
fig, ax = plt.subplots()
for key, group in data.groupby(by='origin'):
    group.plot.scatter('mpg', 'disp', c=colors[key], label=key, ax=ax);
```

## Seaborn 
>Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

http://seaborn.pydata.org/index.html

Seaborn is usually imported as `sns`

```python
import seaborn as sns
```

Let's re-create the histograms by gender with seaborn with the figure level `displot()` function.

```python
# Use gender to split age into columns
sns.displot(x='mpg', col='origin', data=data);
```

We can display the counts in the same plot, one on top of the other.

```python
# Use origin to color (hue) in the same plot
sns.displot(x='mpg', hue='origin', data=data);
```

To have an idea of the split between origins, we can stack the counts, adding up to total.

```python
sns.displot(x='mpg', hue='origin', data=data, multiple='stack');
```

We can look at the differences in mpg with a boxplot too

```python
sns.boxplot(x='origin', y='mpg', data=data);
```

Let's re-create the scatter plot to see if mpg and displacement (disp) are correlated by origin.

To make the legend show strings we will create a origin string column with region-specific strings rather than 1,2 and 3.

```python
data['origin_str'] = data['origin'].replace([1, 2, 3], ['region-1', 'region-2', 'region-3'])
```

```python
ax = sns.scatterplot(x='mpg', y='disp', data=data, hue='origin_str')
```

Adding a regression line helps with visualizing the relationship

```python
ax = sns.lmplot(x='mpg', y='disp', data=data, hue='origin_str')
```

Maybe there are other correlations in the data set. Pairplot is a great way to get an overview

```python
sns.pairplot(data, vars=['mpg', 'disp', 'accel', 'origin', 'weight'], hue='origin_str');
```

As an alternative, we can visualize the correlation matrix as a heatmap

```python
g = sns.heatmap(data[['mpg', 'disp', 'accel', 'weight']].corr(method='spearman'), 
            annot=True)
```
